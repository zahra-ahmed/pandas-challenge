```python
import pandas as pd
```

```python
# file path
file = "purchase_data.csv"
```

```python
# read file
purchase_data = pd.read_csv(file)
```

```python
# display data
purchase_data.head()
```

```python
#PLAYER COUNT

# calculating and displaying total number of players in a DF

total_players = len(purchase_data["SN"].unique())
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

```python
#PURCHASING ANALYSIS (TOTAL)

# calculating and storing purchasing totals in DF

total_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
num_purchase = purchase_data["Item Name"].count()
total_rev = purchase_data["Price"].sum()

purchasing_total_df = pd.DataFrame({"Number of Unique Items" : [total_items],
                                    "Average Price" : [avg_price],
                                    "Number of Purchases" : [num_purchase],
                                    "Total Revenue" : [total_rev]})
```

```python
# formatting and displaying purchasing total DF

purchasing_total_df["Average Price"] = purchasing_total_df["Average Price"].map("${:.2f}".format)
purchasing_total_df["Total Revenue"] = purchasing_total_df["Total Revenue"].map("${:.2f}".format)

purchasing_total_df
```

```python
# GENDER DEMOGRAPHICS

# drop duplicate SNs to get gender and age demographics

unique_sn = purchase_data.drop_duplicates(subset="SN", keep='first')
```

```python
# calculating and storing gender demographics in DF

gender_totals = unique_sn["Gender"].value_counts()
gender_percent = ((unique_sn["Gender"].value_counts()) / total_players * 100)

gender_dem_df = pd.DataFrame({"Total Count": gender_totals,
                             "Percentage of Players": gender_percent})
```

```python
# formatting and displaying gender demographics DF

gender_dem_df["Percentage of Players"] = gender_dem_df["Percentage of Players"].map("{:.2f}%".format)

gender_dem_df
```

```python
# PURCHASING ANALYSIS (GENDER)

# group origianal data by gender for purchasing analysis

gender_group = purchase_data.groupby(['Gender'])
gender_group.count().head()
```

```python
# calculations on gender purchasing

gen_purchase_count = gender_group["Purchase ID"].count()
gen_avg_price = (gender_group["Price"].mean())
gen_total_value = (gender_group["Price"].sum())
gen_avg_purchase = (gender_group["Price"].sum() / gender_totals)
```

```python
# storing gender purchasing data in DF

gen_purchasing_df = pd.DataFrame({"Purchase Count": gen_purchase_count,
                                "Average Purchase Price": gen_avg_price,
                                "Total Purchase Value": gen_total_value,
                                "Avg Total Purchase per Person": gen_avg_purchase})
```

```python
# formatting and displaying gender purchasing data

gen_purchasing_df["Average Purchase Price"] = gen_purchasing_df["Average Purchase Price"].map("${:.2f}".format)
gen_purchasing_df["Total Purchase Value"] = gen_purchasing_df["Total Purchase Value"].map("${:.2f}".format)
gen_purchasing_df["Avg Total Purchase per Person"] = gen_purchasing_df["Avg Total Purchase per Person"].map("${:.2f}".format)

gen_purchasing_df
```

```python
# AGE DEMOGRAPHICS

# cut data and group into bins for age demographics (without duplicate SNs)

age_bin = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

unique_sn["Age Range"] = pd.cut(unique_sn["Age"], age_bin, labels=age_bin_labels)
age_data = unique_sn.groupby("Age Range")
age_data.count().head()
```

```python
# calculate age demoghraphics

age_totals = age_data["SN"].count()
age_percent = ((age_data["SN"].count()) / total_players * 100)
```

```python
# store age demographics in DF

age_dem_df = pd.DataFrame({"Total Count": age_totals,
                             "Percentage of Players": age_percent})
```

```python
# format and display age demographics

age_dem_df["Percentage of Players"] = age_dem_df["Percentage of Players"].map("{:.2f}%".format)

age_dem_df
```

```python
# PURCHASING ANALYSIS (AGE)

# cut data and group into bins for age purchasing data (with duplicate SNs)

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], age_bin, labels=age_bin_labels)
age_purchase_data = purchase_data.groupby("Age Range")
age_purchase_data.count().head()
```

```python
# calculations for age purchasing analysis

age_purchase_count = age_purchase_data["Purchase ID"].count()
age_avg_price = (age_purchase_data["Price"].mean())
age_total_value = (age_purchase_data["Price"].sum())
age_avg_purchase = (age_purchase_data["Price"].sum() / age_totals)
```

```python
# storing age purchasing data in DF

age_purchasing_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                                "Average Purchase Price": age_avg_price,
                                "Total Purchase Value": age_total_value,
                                "Avg Total Purchase per Person": age_avg_purchase})
```

```python
# formatting and displaying age purchasing data

age_purchasing_df["Average Purchase Price"] = age_purchasing_df["Average Purchase Price"].map("${:.2f}".format)
age_purchasing_df["Total Purchase Value"] = age_purchasing_df["Total Purchase Value"].map("${:.2f}".format)
age_purchasing_df["Avg Total Purchase per Person"] = age_purchasing_df["Avg Total Purchase per Person"].map("${:.2f}".format)

age_purchasing_df
```

```python
# TOP SPENDERS

# group by SN for top spender

top_group = purchase_data.groupby("SN")
```

```python
# calculate values for top spender

top_purchase_count = top_group["Purchase ID"].count()
top_avg_price = top_group["Price"].mean()
top_total_value = top_group["Price"].sum()
```

```python
# storing top spender analysis in DF in descending order

top_group_df = pd.DataFrame({"Purchase Count": top_purchase_count,
                                "Average Purchase Price": top_avg_price,
                                "Total Purchase Value": top_total_value})

top_group_df = top_group_df.sort_values(["Total Purchase Value"], ascending=False)
```

```python
# formatting and displaying top spender data

top_group_df["Average Purchase Price"] = top_group_df["Average Purchase Price"].map("${:.2f}".format)
top_group_df["Total Purchase Value"] = top_group_df["Total Purchase Value"].map("${:.2f}".format)

top_group_df.head()
```

```python
# MOST POPULAR ITEMS

# retrieve and group by items

pop_items = purchase_data[["Item ID", "Item Name", "Price", "Purchase ID"]]
pop_items = pop_items.groupby(["Item ID", "Item Name"])
pop_items.count().head()
```

```python
# calculate values for most popular item

item_purchase_count = pop_items["Purchase ID"].count()
item_price = pop_items["Price"].mean()
item_total_value = pop_items["Price"].sum()
```

```python
# storing most popular items in DF in descending order

item_group_df = pd.DataFrame({"Purchase Count": item_purchase_count,
                                "Item Price": item_price,
                                "Total Purchase Value": item_total_value})

most_pop = item_group_df.sort_values(["Purchase Count"], ascending=False)
```

```python
# formatting and displaying most popular item

most_pop["Item Price"] = most_pop["Item Price"].map("${:.2f}".format)
most_pop["Total Purchase Value"] = most_pop["Total Purchase Value"].map("${:.2f}".format)

most_pop.head()
```

```python
# MOST PROFITABLE ITEMS

# sorting items by descending total purcahse value

most_profit = item_group_df.sort_values(["Total Purchase Value"], ascending=False)
```

```python
# formatting and displaying most profitable items

most_profit["Item Price"] = most_profit["Item Price"].map("${:.2f}".format)
most_profit["Total Purchase Value"] = most_profit["Total Purchase Value"].map("${:.2f}".format)

most_profit.head()
```
